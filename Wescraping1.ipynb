```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_cities_in_India_by_population'
page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
tables = soup.find_all('table')
```

```python
print(len(tables))
```

```python
print(tables)
```

```python
table = soup.find_all('table')[0]
```

```python
state_titles = table.find_all('th')
```

```python
state_titles
```

```python
import re
state_tiles = [title.text.strip() for title in state_titles ]
clean_columns = [re.sub(r'\[\w+\]', '', col) for col in state_tiles]
clean_columns.pop(4)
clean_columns.insert(0,'id')
print(clean_columns)
```

```python
table_date = table.find_all('tr')
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = clean_columns)
df
```

```python
row_id = 1
for row in table_date[1:]:
  row_data = row.find_all('td')
  individual_row_data = [data.text.strip() for data in row_data ]
  individual_row_data.pop(4)
  individual_row_data.insert(0,row_id)
  row_id = row_id+1


  length = len(df)
  df.loc[length] = individual_row_data
```
