```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_states_and_union_territories_of_India_by_area'
page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
tables = soup.find_all('table')[1]
```

```python
print(tables)
```

```python
state_titles = tables.find_all('th')
```

```python
print(state_titles)
```

```python
import re
statet = [titles.text.strip() for titles in state_titles]
cleaned_column_names = statet[0:4]
print(cleaned_column_names)
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = cleaned_column_names)
```

```python
table_data = tables.find_all('tr')
```

```python
for row in table_data[1:]:
  row_data = row.find_all('td')
  individual_row_data = [data.text.strip() for data in row_data]
  individual = [re.sub(r'\([A-Z][A-Z]?[0-9]+\)','',ro) for ro in individual_row_data]
  inserted_data = individual[0:4]
  length = len(df)
  df.loc[length] = inserted_data
```

```python
df
```

```python
df = df.drop(index=36)
```
